```python
import pandas as pd
Wage=pd.read_csv('Wage.csv', header=0)
#Wage.head(10)
```

```python
#スプライン回帰
import numpy as np
from patsy import dmatrix
basis1=dmatrix("bs(Wage.age, knots=(25,40,60), degree=3)", {
    "Wage.age":Wage.age}, return_type='dataframe')
import statsmodels.api as sm
fit1=sm.GLM(Wage.wage, basis1).fit()
#fit1.summary()
```

```python
#基底関数の値を作成
basis2=dmatrix("bs(Wage.age, df=6)", {"Wage.age":Wage.age}, return_type='dataframe')
fit2=sm.GLM(Wage.wage, basis2).fit()
#fit2.summary()
```

```python
#自然スプライン
basis3=dmatrix("cr(Wage.age, df=4)", {"Wage.age":Wage.age}, return_type='dataframe')
fit3=sm.GLM(Wage.wage, basis3).fit()
#fit3.summary()
```

```python
ag=np.arange(Wage.age.min(), Wage.age.max()).reshape(-1,1)
pred1 = fit1.predict(dmatrix("bs(ag, knots=(25,40,60))",
{"age_grid": ag}, return_type='dataframe'))
pred2 = fit2.predict(dmatrix("bs(ag, df=6)", {"age_grid": ag},
return_type='dataframe'))
pred3 = fit3.predict(dmatrix("cr(ag, df=4)", {"age_grid": ag},
return_type='dataframe'))
```

```python
import matplotlib.pyplot as plt
plt.scatter(Wage.age, Wage.wage, facecolor='None',
edgecolor='k', alpha=0.1)
plt.plot(ag,pred1, color='r',label='Cubic spine with knots at [25, 40, 60]')
plt.plot(ag, pred2, color='g', label='Cubic spine with df=6')
plt.plot(ag, pred3, color='b', label='Natural spline df=4')
plt.legend()
plt.xlabel('age')
plt.ylabel('wage')
plt.show()
```

```python
#一般化加法モデル
age_basis = dmatrix("cr(Wage.age, df=5)", {"Wage.age": Wage.age},return_type='dataframe')
year_basis = dmatrix("cr(Wage.year, df=4)", {"Wage.year":Wage.year}, return_type='dataframe').drop (['Intercept'], axis = 1)
dummies = pd.get_dummies(Wage.education)
dummies = dummies.drop(dummies.columns[0], axis = 1)
x_all = pd.concat([age_basis, year_basis, dummies], axis = 1)

gam1_fit = sm.OLS(Wage.wage, x_all).fit()
gam1_fit.summary()
```

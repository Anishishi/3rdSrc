```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
pi = math.pi
x0 = np.linspace(0,0.004,44100/250)
y0 = np.sin(2*pi*3582*x0)
x1 = np.linspace(0,0.004, 4410/250)
y1 = np.sin(2*pi*3528*x1)
x2 = np.linspace(0,0.004, 7056/250)
y2 = np.sin(2*pi*3528*x2)

plt.plot(x0,y0,label='f=44100Hz')
plt.plot(x1,y1,label='f=4410Hz')
plt.plot(x2, y2, label='f=7056Hz')
plt.title('Three Sin Graghs(f0=3528Hz)')
plt.xlabel('time[s]')
plt.ylabel('amplitude(max=1.0)')
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0, fontsize=10)

plt.savefig('two_sin_graghs.png')
```

```python
pi = math.pi
x0 = np.arange(0.002,0.007,1/44100)
y0 = np.sin(2*pi*3582*x0)
x1 = np.arange(0.002,0.007, 1/4410)
y1 = np.sin(2*pi*3528*x1)
x2 = np.arange(0.002,0.007,1/7056)
y2 = np.sin(2*pi*3528*x2)

plt.plot(x0,y0,label='f=44100Hz')
plt.plot(x1,y1,label='f=4410Hz')
plt.plot(x2, y2, label='f=7056Hz')
plt.title('Three Sin Graghs(f0=3528Hz)')
plt.xlabel('time[s]')
plt.ylabel('amplitude(max=1.0)')
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0, fontsize=10)
```

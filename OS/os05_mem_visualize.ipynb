```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

## record_page_faults_1の出力(経過時間)の表示

* show_mb_vs_t(filename) は filename (record_page_faults_1 の出力)の,
 - 第0列目(アクセスしたページ数をMBに直したもの)をx軸に, 
 - 第1列目(経過時間)をy軸に
とって可視化する

```python
def show_mb_vs_t(filenames):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    for filename in filenames:
        log = np.loadtxt(filename)
        accessed = log[:,0] / (1024 * 1024)
        elapsed  = log[:,1]
        ax.plot(accessed, elapsed)
    plt.show()
```

触った領域 vs. 経過時間 (150MBの領域を3回アクセス)

```python
show_mb_vs_t(["time150x3.log"])
```

触った領域 vs. 経過時間 (300MBの領域を3回アクセス)

```python
show_mb_vs_t(["time300x3.log"])
```

触った領域 vs. 経過時間 (450MBの領域を3回アクセス)

```python
show_mb_vs_t(["time450x3.log"])
```

触った領域 vs. 経過時間 (3ケースを重ねて表示)

```python
show_mb_vs_t(["time150x3.log", "time300x3.log", "time450x3.log"])
```
